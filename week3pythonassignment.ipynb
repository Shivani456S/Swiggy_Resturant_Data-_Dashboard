##Delivery Time Outliers:
Identify any outliers in delivery time using a box plot and explain what these outliers might indicate about those restaurants

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.boxplot(x=df['Delivery time'] , color= 'yellow' ,fliersize=7, linewidth=1.4)
plt.title("Delivery Time with the Outliers")
plt.xlabel('Delivery time (min)')
plt.tight_layout()
plt.show()
```

```python
# Calculate the IQR for 'Delivery Time' to identify outliers
Q1 = df['Delivery time'].quantile(0.25)
Q3 = df['Delivery time'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
outliers = df[(df['Delivery time'] < lower_bound) | (df['Delivery time'] > upper_bound)]

print("Outliers in delivery time:")
print(outliers[['Restaurant', 'Delivery time']])
```

##Grouped Analysis:
Perform a grouped analysis to determine the average price and average rating for each city and visualize the results with a grouped bar chart.

```python
grouped_analysis = df.groupby('City')[['Price', 'Avg ratings']].mean().reset_index()
print(grouped_analysis)
```

```python
plt.figure(figsize=(5, 4))
bar_width = 0.35
index = np.arange(len(grouped_analysis))
bar1 = plt.bar(index, grouped_analysis['Price'], bar_width, label='Average Price', color='yellow', edgecolor='green')
bar2 = plt.bar(index + bar_width, grouped_analysis['Avg ratings'], bar_width, label='Average Rating', color='green')
plt.xlabel('City')
plt.ylabel('Value')
plt.title('Average Price and Rating for Each City')
plt.xticks(index + bar_width / 2, grouped_analysis['City'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
